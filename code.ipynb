## ***3.1 Xây dựng model***

* ***Adam (Adaptive Moment Estimation)*** là một phương pháp tổng hợp Adagrad và RMSprop để cập nhật tốc độ học. Adam sử dụng trung bình bình phương của gradient như RMSprop và tích lũy gradient bình phương như Adagrad
* ***SGD (Stochastic Gradient Descent)*** là thuật toán tối ưu hóa cơ bản theo họ gradient. Thuật toán này rất triển khai, có nền tảng lý thuyết vững chắc, cực kỳ ổn định trong quá trình huấn luyện, kết quả đạt được có thể so sánh với các thuật toán khác. Ý tưởng của thuật toán khá đơn giản, đó là “tính giá trị gradient của mỗi tham số, và đi một bước nhỏ theo chiều của gradient”. Nếu chúng ta lặp đi lặp lại quá trình này, và ngẫu nhiên chọn (stochastic) một tập batch trong tập huấn luyện, mô hình chúng ta sẽ được cải tiến dần đến đểm hội tụ.
* ***Adagrad (Adaptive Gradient)*** là một kỹ thuật học máy tiên tiến, thực hiện giảm dần độ dốc bằng cách thay đổi tốc độ học tập. Adagrad được cải thiện hơn bằng cách cho trọng số học tập chính xác dựa vào đầu vào trước nó để tự điều chỉnh tỉ lệ học theo hướng tối ưu nhất thay vì với một tỉ lệ học duy nhất cho tất cả các nút
* ***RMSProp (Root Mean Square Propagation)*** sử dụng trung bình bình phương của gradient để chuẩn hóa gradient. Có tác dụng cân bằng kích thước bước - giảm bước cho độ dốc lớn để tránh hiện tượng phát nổ độ dốc (Exploding Gradient), và tăng bước cho độ dốc nhỏ để tránh biến mất độ dốc (Vanishing Gradient). RMSProp tự động điều chỉnh tốc độ học tập, và chọn một tỉ lệ học tập khác nhau cho mỗi tham số.
* ***Nadam (Nesterov-accelerated Adaptive Moment Estimation)*** là một biến thể của Adam và SGD. Thuật toán này kết hợp giữa Momentum và Adam để cải thiện việc tìm kiếm bộ trọng số tốt nhất. Nadam có thể giúp giảm thiểu hiện tượng overshooting và tăng tốc quá trình học.

```python
from keras.layers import Dense
from keras.models import Sequential
from keras.optimizers import Adam, SGD, Adagrad, RMSprop, Nadam

def create_model(opt):
    model = Sequential()
    model.add(Dense(128, activation='relu', input_shape=(X.shape[1],)))
    model.add(Dense(64, activation='relu'))
    model.add(Dense(32, activation='relu'))
    model.add(Dense(2, activation='sigmoid'))
    model.compile(loss='binary_crossentropy', optimizer=opt, metrics=['accuracy'])
    model.summary()

    return model
```

```python
models = {
    'Adam': create_model(Adam()),
    'SGD': create_model(SGD()),
    'Adagrad': create_model(Adagrad()),
    'RMSprop': create_model(RMSprop()),
    'Nadam': create_model(Nadam()),
}
```

## ***3.2 Tạo hàm đánh giá model***

```python
from sklearn.metrics import accuracy_score, f1_score, classification_report
import numpy as np

def evaluate_model(model):
    start = time.time()                 # thời gian bắt đầu
    y_pred = model.predict(X_test)
    time_test = time.time() - start     # tính thời gian quá trình dự đoán

    y_pred = np.argmax(y_pred, axis=1)

    accuracy = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred, average='weighted')
    print(classification_report(y_test, y_pred))
    
    return (time_test, accuracy, f1)
```

## ***3.3 Huấn luyện mô hình***

```python
import time
compare = {}
for name, model in models.items():
    print(f'Model {name}')
    # training
    start = time.time()
    history = model.fit(X_train, y_train, epochs=30, batch_size=64, validation_data=(X_validation, y_validation))
    time_train = time.time() - start

    # testing
    time_test, accuracy, f1 = evaluate_model(model)

    # save compare
    compare[name] = {
        'time_train': time_train,
        'time_test': time_test,
        'accuracy': accuracy,
        'f1': f1,
        'history': history
        }
    print('-'*75)
```

# **4. Đánh giá các mô hình**

## ***4.1 Đồ thị history thu được từ quá trình huấn luyện***

```python
import matplotlib.pyplot as plt

def plot_(title, values):
    plt.title(title)
    for name in compare.keys():
        plt.plot(compare[name]['history'].history[values], label=f'{name}')
    plt.legend()

plt.figure(figsize=(20,10))
plt.subplot(2,2,1)
plot_('Train accuracy', 'accuracy')
plt.subplot(2,2,2)
plot_('Validation accuracy', 'val_accuracy')
plt.subplot(2,2,3)
plot_('Train loss', 'loss')
plt.subplot(2,2,4)
plot_('Validation loss', 'val_loss')
plt.show()
```
